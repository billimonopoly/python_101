```python
#################
#Luhn's Algorithm
#################

#this function is used in step 3, in order to add the digits of even index elements over 10
def getSum(n): 
    sum = 0
    for digit in str(n): 
      sum += int(digit)      
    return sum

def is_credit(credit_card_number):
  #this determines whether the number should be proceeded with
  if credit_card_number.isnumeric():
    #this seperates the series of numbers into a list of integers
    credit_card_list = [int(x) for item,x in enumerate(str(credit_card_number))]
    #this takes the even index elements
    step1 = credit_card_list[-2::-2]
    #this multiplies each element from step 1 by 2
    step2 = [i * 2 for i in step1]
    #this adds the digits of items that are greater than 10, and replace the original element in step 2
    step3 = [getSum(item) if item >= 10 else item for item in step2]
    #this once again adds the altered even index elements to the odd index elements
    #here order doesn't matter since they will all be added together
    step4 = step3 + credit_card_list[-1::-2]
    #this adds the elements of the list together
    step5 = sum(step4)
    #this determines the remainder of the added list
    step6 = step5 % 10
    if step6 == 0:
        return True
    else:
        return False
  else:
      return False
      
##############################
#is_credit Function Test Cases
##############################

#Test Case 1
if is_credit("6011000000000004") == True:
  print('pass')
else:
  print('fail')

#Test Case 2
if is_credit("79927398713") == True:
  print('pass')
else:
  print('fail')

#Test Case 3
if is_credit("30000000000004") == True:
  print('pass')
else:
  print('fail')

#Test Case 4
if is_credit("6911600873502604") == False:
  print('pass')
else:
  print('fail')

#Test Case 5
if is_credit("1234567890") == False:
  print('pass')
else:
  print('fail')

def is_visa(credit_card_number):
  #this checks if the input is valid
  if credit_card_number.isnumeric():
    if is_credit(credit_card_number) == True:
      #this turns the card number into a list
      credit_card_list = [int(x) for item,x in enumerate(str(credit_card_number))]
      #this. determines the length of the card number
      len_credit_card_list = len(credit_card_list)
      #this determines thhe first number of the card
      first_number = credit_card_list[0]
      #if the card number begins with 4 and has either 13 or 16 digits, then it is a potential visa card number
      if first_number == 4:
        if len_credit_card_list == 13 or len_credit_card_list == 16:
          return True
        else:
          return False
      else:
        return False
    else:
      return False
  else:
    return False
    
def is_mastercard(credit_card_number):
  #this checks if the input is valid
  if credit_card_number.isnumeric():
    if is_credit(credit_card_number) == True:
      #this turns the card number into a list
      credit_card_list = [int(x) for item,x in enumerate(str(credit_card_number))]
      #this. determines the length of the card number
      len_credit_card_list = len(credit_card_list)
      #this determines thhe first number of the card
      first_number = credit_card_list[0]
      #if the card number begins with 5 and has 16 digits, then it is a potential mastercard card number
      if first_number == 5:
        if len_credit_card_list == 16:
          return True
        else:
          return False
      else:
        return False
    else:
      return False
  else:
    return False

def main(input):
  #this allows the function to take 1 input, and split it by the spaces
  x=input.split()
  #this ensures the correct format is used
  if len(x) == 2:
    #this defines the 2 elements of the list as card type and credit card number 
    card_type=x[0]
    credit_card_number=x[1]
    #this makes the card type case insensitive
    card_type = card_type.lower()
    if card_type == 'visa':
      if is_visa(credit_card_number) == True:
        print('Valid Visa Number')
        return True
      if is_visa(credit_card_number) == False:
        print('Invalid Visa Number')
        return False
    elif card_type == 'mastercard':
      if is_mastercard(credit_card_number) == True:
        print ('Valid MasterCard Number')
        return True
      else:
        print('Invalid MasterCard Number')
    else:
      print('Invalid Card Type')
      return False
  else:
    print('ERROR: Correct fomat was not used')
    return False

#########################
#Main Function Test Cases
#########################

#Test Case 1
if main('Visa 4111111111111111') == True:
  print('pass')
else:
  print('fail')

#Test Case 2
if main('visA 4111111111111111') == True:
  print('pass')
else:
  print('fail')

#Test Case 3
if main(' mastercard 5500000000000004') == True:
  print('pass')
else:
  print('fail')

#Test Case 4
if main(' mastercard 123ACBCDEG4') == False:
  print('pass')
else:
  print('fail')

#Test Case 5
if main(' AmericanExpress 340000000000009') == False:
  print('pass')
else:
  print('fail')
```
